# Test and Train

```python
#split into testing and training
X = pd.get_dummies(features_df, columns = ['age', 'work_class', 'education', 'marital_status', 
                                           'occupation', 'relationship', 'race', 'sex','hours_per_week', 'native_country'])
y = target_df['salary']
X.head()
```

```python
y.value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
```

# SMOTE Oversampling

```python
#Import Packages
from imblearn.over_sampling import SMOTE
from imblearn.over_sampling import RandomOverSampler
from collections import Counter
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import balanced_accuracy_score
# Resample the training data with SMOTE
# Instantiate the model
smote = SMOTE(random_state=1)
# Resample the targets
X_resampled, y_resampled = smote.fit_resample(X_train, y_train)
Counter(y_resampled)
```

```python
# Train the Logistic Regression model using the resampled data
logreg = LogisticRegression(solver='lbfgs', random_state=1)
logreg.fit(X_resampled, y_resampled)
```

```python
# Calculated the balanced accuracy score
y_pred = logreg.predict(X_test)
balanced_accuracy_score(y_test, y_pred)
```

# Random Over Sampling 

```python
# Instantiate the model
ros = RandomOverSampler(random_state=1)
# Resample the targets
X_resampled, y_resampled = ros.fit_resample(X_train, y_train)
Counter(y_resampled)
```

```python
# Train the Logistic Regression model using the resampled data
logreg = LogisticRegression(solver='lbfgs', random_state=1)

# Fit
logreg.fit(X_resampled, y_resampled)
```

```python
# Calculated the balanced accuracy score
y_pred = logreg.predict(X_test)
balanced_accuracy_score(y_test, y_pred)
```

# Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
```

```python
# Display the confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Create a DataFrame from the confusion matrix.
cm_df = pd.DataFrame(
    cm, index=["Actual Over_50k", "Actual Under_50k"], columns=["Predicted Over_50k", "Predicted Under_50k"])
cm_df
```
