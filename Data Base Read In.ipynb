```python
## Create Features DataFrame
features_df = project_df.filter(['age', 'workclass', 'education', 'marital-status', 'occupation', 'relationship', 'race', 'sex','hours-per-week', 'native-country' ], axis = 1)
features_df
```

```python
#create target df
target_df = project_df.filter(['age','salary'], axis=1)
target_df
```

```python
# Renaming titles columns
features_df = features_df.rename(columns={'workclass':'work_class','marital-status': 'marital_status','hours-per-week':'hours_per_week','native-country':'native_country'})
features_df
```

```python
# Add this import for sql database
import sqlite3
```

```python
# connecting to 'features.db'
conn = sqlite3.connect('Salary.db')
```

```python
# Adding feature_df database
features_df.to_sql(name ='features', con=conn, if_exists='replace', index=False)
conn.commit
```

```python
# Adding target_df to database
target_df.to_sql(name ='target', con=conn, if_exists='replace', index=False)
conn.commit
```
